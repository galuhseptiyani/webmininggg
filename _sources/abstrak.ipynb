```python
url = 'https://pta.trunojoyo.ac.id/c_search/byprod/10/'
s = HTMLSession()
```

function save untuk menyimpan data dalam bentuk csv

```python
def save():
  df = pd.DataFrame(ta_list)
  df.to_csv('data_abstrak_tif.csv')
  print('Telah tersimpan file csv')
```

```python
hal = 0
ta_list=[]

for page in range(1, 20):
    hal += 1
    #print('Halaman ', hal)
    
    r = s.get(url+str(page))
    req = r.content
    link_text = []
    soup = BeautifulSoup(req, 'html.parser')

    for a in soup.findAll('a', attrs={'class':'gray button'}, href=True):
        li = a['href']
        link_text.append(li)

    for item in link_text:
        r = s.get(item)
        title = r.html.find('a.title', first=True).text.strip()
        penulis = r.html.xpath('//*[@style="padding:2px 2px 2px 2px;"]', first=True).text.strip()
        #dosen1 = r.html.xpath('//*[@style="padding:2px 2px 2px 2px;"]', first=True).text

        abstrak = r.html.xpath('//*[@align="justify"]', first=True).text.strip()

        detail = {
            'Judul' : title,
            'Penulis' : penulis,
            'Link' : item,
            'Abstrak' : abstrak
        }

        ta_list.append(detail)
        #output()
save()
```
